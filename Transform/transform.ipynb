```python
import glob                    
import pandas as pd 
import copy  
```

```python
path_teams_files = '../Outputs/teams'
extracted_data = []
```

```python
def extract_from_csv(file_to_process):
    dataframe = pd.read_csv(file_to_process)
    return dataframe
```

```python
columns=['team','goals_scored','goals_conceded','own_goals', 'foul_goals',
'penalty_goals','goals_inside_the_area','goal_from_outside_the_area', 
'finishing_on_the_post', 'submissions_out','fouls_received', 
'fouls_committed', 'offsides','yellow_cards', 'red_cards', 
'tough_defenses','penalty_saves', 'ball_steals','wrong_passes']
```

```python
def transform():
  
  dic = {}
  
  for csvfile in glob.glob(f"{path_teams_files}/*.csv"):

    current_csv = extract_from_csv(csvfile)
    current_columns = list(current_csv.columns)

    for index, row in current_csv.iterrows():

      value1 = int(row[1])
      value2 = int(row[2])

      try:

        aux_dic = dic[row[0]]
    
        if not (current_columns[1] in aux_dic):
          aux_dic[current_columns[1]] = value1
        
        if not (current_columns[2] in aux_dic):
          aux_dic[current_columns[2]] = value2
         
      except:

        aux_dic = {}
        aux_dic[current_columns[1]] = value1
        aux_dic[current_columns[2]] = value2
        dic[row[0]] = aux_dic
  

  aux_list = []

  for key, value in dic.items():
    new_dic = copy.deepcopy(value)
    new_dic['team'] = key
    aux_list.append(new_dic)

  ndf = pd.DataFrame(aux_list, columns=columns)
  return ndf
```
